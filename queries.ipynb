### Combo de python, pandas e sqlite3

```python
import pandas as pd
import sqlalchemy
import sqlite3
```

```python
df = pd.read_csv('sales_data_simplified.csv')
```

```python
conn = sqlite3.connect('enterprise.db')
```

```python
df.to_sql(name='sales', con=conn)
```

```python
!sqlite3 enterprise.db .tables
```

```python
# !sqlite3 enterprise.db PRAGMA table_info(sales)
```

```python
df_read = pd.read_sql('SELECT * FROM sales', con=conn)
```

```python
df_read
```

```python
query = """
        SELECT "id", "product", "quantity", "price"
        FROM sales
        """
df_read = pd.read_sql(query, con=conn, index_col="id")
df_read
```

```python
query = """
        SELECT *
        FROM sales
        WHERE "product" = "USB-C Charging Cable"
        """
df_read = pd.read_sql(query, con=conn, index_col="id")
df_read
```

```python
query = """
        SELECT *, "price"*"quantity" as 'total'
        FROM sales
        
        """
df_read = pd.read_sql(query, con=conn, index_col="id")
df_read
```

```python
df_read.to_sql(name='sales', con=conn, if_exists='replace')
```

```python
df_read = pd.read_sql('SELECT * FROM sales', con=conn)
df_read
```
